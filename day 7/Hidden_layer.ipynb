```python
from tensorflow import keras
from tensorflow.keras.layers import Input, Dense, Reshape, Flatten
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.optimizers import Adam,Adadelta
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
x_train=x_train/255
```

```python
x_test=x_test/255
```

```python
x_train_flatten=x_train.reshape(len(x_train),28*28)
```

```python
x_test_flatten=x_test.reshape(len(x_test),28*28)
```

```python
model=keras.Sequential([
    keras.layers.Dense(50, activation='relu', input_shape=(784,)),
    keras.layers.Dense(10, activation='softmax'),
])
model.compile(optimizer=Adam(learning_rate=0.05), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
df=model.fit(x_train_flatten,y_train,epochs=20,batch_size=1000,validation_data=(x_test_flatten,y_test))
```

```python
plt.plot(df.history['loss'])
plt.plot(df.history['val_loss'])
plt.title('Training and validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['loss', 'val_loss'], loc='upper left')
plt.show()
```

```python
plt.plot(df.history['accuracy'])
plt.plot(df.history['val_accuracy'])
plt.title('Training and validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['accuracy', 'val_accuracy'], loc='upper left')
plt.show()
```
