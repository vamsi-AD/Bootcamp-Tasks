```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
#iris_df['target'] = iris.target
```

```python
iris_df
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(iris_df)
scaled_df = pd.DataFrame(scaled_features, columns=iris.feature_names)
```

```python
scaled_df
```

```python
km = KMeans(n_clusters=3, random_state=42, n_init='auto')
scaled_df['cluster'] = km.fit_predict(scaled_features)
scaled_df['target'] = iris.target
```

```python
scaled_df.describe()
```

```python
cluster_mean= scaled_df.groupby('cluster')[iris.feature_names].mean()
cluster_counts = scaled_df['cluster'].value_counts().sort_index()
```

```python
target_mean= scaled_df.groupby('target')[iris.feature_names].mean()
target_counts = scaled_df['target'].value_counts().sort_index()
```

```python
print(cluster_mean, target_mean)
```
