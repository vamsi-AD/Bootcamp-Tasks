```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import numpy as np
```

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
iris_df.head()
```

```python
X = iris_df[['petal length (cm)','petal width (cm)']]
```

```python
X
```

```python
def plot_kmeans_with_initial_plots(X,initial_plots, title):
    kmeans = KMeans(n_clusters=3, init=initial_plots, n_init=1)
    iris_df['clusters'] = kmeans.fit_predict(X)
    
    plt.scatter(X['petal length (cm)'],X['petal width (cm)'], c=iris_df['clusters'], cmap='viridis')
    plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s=300, c='red', marker='*', label='centroids')
    plt.title(title)
    plt.xlabel('petal length (cm)')
    plt.ylabel('petal width (cm)')
    plt.legend()
    plt.show()
```

```python
kmeans_random = KMeans(n_clusters=3, random_state=42)
iris_df['cluster_random'] = kmeans_random.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X['petal length (cm)'], X['petal width (cm)'], c=iris_df['cluster_random'], cmap='viridis')
plt.scatter(kmeans_random.cluster_centers_[:, 0], kmeans_random.cluster_centers_[:, 1], s=100, c='red', marker='o', label='Centroids')
plt.title("K-means with Random Initial Centroids")
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend()
plt.show()
```

```python
initial_centroids_1 = np.array([[1.5, 0.2], [5.0, 1.5], [6.5, 2.0]])
plot_kmeans_with_initial_plots(X, initial_centroids_1, "K-means with Custom Initial Centroids 1")
```

```python
initial_centroids_2 = np.array([[6.0, 0.5], [4.0, 2.0], [2.0, 1.5]])
plot_kmeans_with_initial_plots(X, initial_centroids_2, "K-means with Custom Initial Centroids 2")
```
