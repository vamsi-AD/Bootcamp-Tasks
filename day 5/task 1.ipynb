```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
import pandas as pd
```

```python
iris = load_iris()
df = pd.DataFrame(data = iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```python
df.shape
```

```python
df
```

```python
df['species'] = df['target'].apply(lambda x: iris.target_names[x])
```

```python
df = df.drop(columns=['species'])
df = df.drop(columns=['target'])
```

```python
data = df
```

```python
k = 4
km = KMeans(n_clusters=k)
km.fit(data)
```

```python
km.inertia_
```

```python
km.cluster_centers_
```

```python
pred = km.fit_predict(data)
```

```python
pred
```

```python
output = data
output['predicted'] = pred
```

```python
output = output.drop(columns=['predicted'])
```

```python
output.head()
```

```python
inertias = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit(data)
    inertias.append(km.inertia_)
```

```python
inertias
```

```python
plt.plot(range(1,11),inertias, marker = 'o')
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(df[iris.feature_names])
df['pca1'] = pca_result[:, 0]
df['pca2'] = pca_result[:, 1]

kmeans_pca = KMeans(n_clusters=3, random_state= 42)
df['cluster_pca'] = kmeans_pca.fit_predict(df[['pca1', 'pca2']])

plt.figure(figsize=(8, 6))
plt.scatter(df['pca1'], df['pca2'], c=df['cluster_pca'], cmap='viridis')
plt.scatter(kmeans_pca.cluster_centers_[:, 0], kmeans_pca.cluster_centers_[:, 1], s=300, c='red', marker='*', label='Centroids')
plt.title('K-means Clustering of Iris Data (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
